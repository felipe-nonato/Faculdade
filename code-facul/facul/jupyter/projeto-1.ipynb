Sim, é possivel trazer os dados diretamente do site, mas para isso precisaremos importar a lib zipfile, já que os dados disponibilizados pela PRF foram liberados de forma comprimida (.zip)

```python
#Importando bibliotecas necessarias
import zipfile as zf
import pandas as pd
import numpy as np

#importando dataframes do site da prf e guardando
df_2018 = pd.read_csv('https://arquivos.prf.gov.br/arquivos/index.php/s/MaC6cieXSFACNWT/download',encoding = 'latin1',compression = 'zip',sep=';')
df_2019 = pd.read_csv('https://arquivos.prf.gov.br/arquivos/index.php/s/kRBUylqz6DyQznN/download',encoding = 'latin1',compression = 'zip',sep=';')
df_2020 = pd.read_csv('https://arquivos.prf.gov.br/arquivos/index.php/s/jdDLrQIf33xXSCe/download',encoding = 'latin1',compression = 'zip',sep=';')
df_2021 = pd.read_csv('https://arquivos.prf.gov.br/arquivos/index.php/s/n1T3lymvIdDOzzb/download',encoding = 'latin1',compression = 'zip',sep=';')
```
