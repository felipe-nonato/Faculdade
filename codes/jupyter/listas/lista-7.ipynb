<h2>Para o Brasil</h2>

```python
#criando um dataframe com os dados da pandemia
df_pandemia = pd.concat([df_2018,df_2019,df_2020,df_2021], keys = ['2018','2019','2020','2021'],names=['ano','index'])
df_pandemia = df_pandemia.reset_index().drop(['index'],axis=1)

#agrupando o dataframe por ano e contando o número de acidentes que aconteceram em cada ano
df_pandemia = df_pandemia.groupby(['ano']).count().iloc[:,:1].rename(columns={'id':'n° acidentes'})

#criando dataframe com os dados de antes da pandemia
before = pd.DataFrame(df_pandemia.iloc[:2,:].sum()).rename(index={'n° acidentes':'antes da pandemia'})

#criando dataframe com os dados de depois da pandemia
after = pd.DataFrame(df_pandemia.iloc[2:,:].sum()).rename(index={'n° acidentes':'durante a pandemia'})

#concatenando os dados de antes e depois da pandemia para comparar
pd.concat([before,after]).rename(columns={0:'n° acidentes'})
```

Percebemos que o numero de acidentes diminuiram durante a pandemia.

```python
df_pandemia
```

Olhando friamente para os números, podemos perceber que em 2020, ano mais afetado pela pandemia, tivemos uma queda drastica do numero de acidentes, uma queda de aproximadamente **8,3%** quando comparado com o ano de 2018.

Porém se olharmos para o ano de 2021, onde começamos a ter a aplicação de vacinas e a volta de algumas atividades presenciais, vemos um aumento de **1,5%** quando comparamos com o ano de 2020. Dessa forma podemos ver um crescimento novamente no número de acidentes.
