```python
#separando dataframes por meio da função groupby, agrupando por unidade federativas,
#e contando o numero de acidentes de cada dia da semana.
uf_18 = df_2018.groupby(['uf']).count().sort_values('id').iloc[:,:1].rename(columns = {"id":"numero de acidentes"})
uf_19 = df_2019.groupby(['uf']).count().sort_values('id').iloc[:,:1].rename(columns = {"id":"numero de acidentes"})
uf_20 = df_2020.groupby(['uf']).count().sort_values('id').iloc[:,:1].rename(columns = {"id":"numero de acidentes"})
uf_21 = df_2021.groupby(['uf']).count().sort_values('id').iloc[:,:1].rename(columns = {"id":"numero de acidentes"})

#juntando dataframes em um só
uf_total = pd.concat([uf_18,uf_19,uf_20,uf_21], keys = ['2018','2019','2020','2021'])

#criando arquivo csv
uf_total.to_csv("Acidentes_Por_UF_[18,19,20,21].csv")

#vizualizando dataframe
uf_total
```

```python
#ordem dos dias da semana em forma de lista
ordem_ds = ["segunda-feira","terça-feira","quarta-feira","quinta-feira","sexta-feira",'sábado','domingo']
#separando pelo grupo dia da semana e contando quantos acidentes cada dia teve

ds_18 = df_2018.groupby(['dia_semana']).count().iloc[:,:1].rename(columns = {"id":"nº acidentes"})
ds_19 = df_2019.groupby(['dia_semana']).count().iloc[:,:1].rename(columns = {"id":"nº acidentes"})
ds_20 = df_2020.groupby(['dia_semana']).count().iloc[:,:1].rename(columns = {"id":"nº acidentes"})
ds_21 = df_2021.groupby(['dia_semana']).count().iloc[:,:1].rename(columns = {"id":"nº acidentes"})

#ordenando os dataframes com base na lista que mostra a ordem correta dos dias da semana

ds_18.index = pd.Categorical(ds_18.index,ordem_ds,ordered=True)
ds_18.sort_index(inplace=True)

ds_19.index = pd.Categorical(ds_19.index,ordem_ds,ordered=True)
ds_19.sort_index(inplace=True)

ds_20.index = pd.Categorical(ds_20.index,ordem_ds,ordered=True)
ds_20.sort_index(inplace=True)

ds_21.index = pd.Categorical(ds_19.index,ordem_ds,ordered=True)
ds_21.sort_index(inplace=True)

#juntando dataframes em um só
ds_total = pd.concat([ds_18,ds_19,ds_20,ds_21], keys = ['2018','2019','2020','2021'])

#criando arquivo csv
ds_total.to_csv("Acidentes_Por_DS_[18,19,20,21].csv")

#vizualizando dataframe
ds_total
```

Para usar apenas o mês precisamos tratar a coluna `data_inversa`. Para isso irei importar a lib `datetime`, da qual a lib `pandas` pode utilizar de alguns artificios para esse tratamento de dados. Com a função `pd.to_datetime()` eu transformo o objeto que era a coluna `data_inversa` em uma data reconhecido pelo script. Com isso feito posso utilizar a função `dt.month` que identifica o mês desta data.

```python
import datetime
```

## Para 2018

```python
df_2018['mes'] = pd.to_datetime(df_2018['data_inversa']).dt.month
```

Para realizar a operação da proporção, pegarei a coluna de `ilesos/feridos_graves` e irei dividir pela coluna `pessoas` (que se refere ao número de pessoas envolvidas no acidente), dessa forma terei o retorno da proporção desejada.

```python
mes_2018 = df_2018.loc[:,['mes','pessoas','ilesos','feridos_graves']]
a = mes_2018.groupby(by='mes').sum()
prop_ilesos = (a['ilesos']/a['pessoas'])
prop_feridos = (a['feridos_graves']/a['pessoas'])
prop = pd.DataFrame(prop_ilesos).rename(columns={0:'proporção ilesos'}).round(4)
prop['proporção feridos graves'] = pd.DataFrame(prop_feridos).round(4)
prop
```

```python
#criando um dataframe com os dados da pandemia
df_pandemia = pd.concat([df_2018,df_2019,df_2020,df_2021], keys = ['2018','2019','2020','2021'],names=['ano','index'])
df_pandemia = df_pandemia.reset_index().drop(['index'],axis=1)

#agrupando o dataframe por ano e contando o número de acidentes que aconteceram em cada ano
df_pandemia = df_pandemia.groupby(['ano']).count().iloc[:,:1].rename(columns={'id':'n° acidentes'})

#criando dataframe com os dados de antes da pandemia
before = pd.DataFrame(df_pandemia.iloc[:2,:].sum()).rename(index={'n° acidentes':'antes da pandemia'})

#criando dataframe com os dados de depois da pandemia
after = pd.DataFrame(df_pandemia.iloc[2:,:].sum()).rename(index={'n° acidentes':'durante a pandemia'})

#concatenando os dados de antes e depois da pandemia para comparar
pd.concat([before,after]).rename(columns={0:'n° acidentes'})
```

Percebemos que o numero de acidentes diminuiram durante a pandemia.

```python
df_pandemia
```

Olhando friamente para os números, podemos perceber que em 2020, ano mais afetado pela pandemia, tivemos uma queda drastica do numero de acidentes, uma queda de aproximadamente **8,3%** quando comparado com o ano de 2018.

Porém se olharmos para o ano de 2021, onde começamos a ter a aplicação de vacinas e a volta de algumas atividades presenciais, vemos um aumento de **1,5%** quando comparamos com o ano de 2020. Dessa forma podemos ver um crescimento novamente no número de acidentes.
