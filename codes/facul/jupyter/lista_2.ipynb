## Listas 
    Usadas normalmente para armazenamento de poucos dados.
    Ex: nomes de alunos de uma sala.
    
    Principais caracteristicas: 

        - mutáveis
        - ordenadas
        - heterogêneas
        - indexada por posição
        - delimitada por colchetes [ ]
        

## Dicionários 
    Usados para dados que precisam de maior organização, onde a indexação pode atrapalhar.
    Ex: Lista telefonica.

    Principais caracteristicas: 
    
        - mutáveis
        - não ordenados
        - heterogêneos
        - indexada por chaves
        - delimitada por chaves { }
        
##  ndarray - Numpy

    Usada como uma lista, mas com facilitação na forma de criar e na possibilidade usar para
    operações com os elementos da lista.
    Ex: Listas de salarios e valor recebido por horas extras que precisam ser somadas.
    
    Principais caracteristicas:
        
        - mutáveis
        - ordenadas
        - "heterogêneas" → (upcasting)
        - indexada por posição
        - delimitada por funções da lib numpy
        
## series - Pandas

    Usada como um DataFrame, é o melhor dos mundos para manipulação de dados, seja por operações,
    ordenamento, filtragem ou troca dos elementos.  
    
    Principais caracteristicas:
    
        - mutáveis
        - ordenadas
        - "heterogêneas" → (upcasting)
        - indexadas da forma que o dev quiser
        - delimitada por funções da lib Pandas

````python
def troca_coluna(df, colA=0, colB=0):
    
    name = df.columns.tolist()
    a, b = name.index(colA), name.index(colB)
    name[b], name[a] = name[a], name[b]
    
    return df[name]

print(troca_coluna(df,'a','c'))
````
